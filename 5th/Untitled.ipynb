```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("insurance_data.csv")
```

```python
x=df[['age']]
y=df['bought_insurance']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=50, test_size=0.2)
```

```python
x
model=LogisticRegression()
model.fit(x_train,y_train)
x_train
```

```python
pred=model.predict(x_test)
print("Predition:",pred)
```

```python
prob=model.predict_proba(x_test)
print("",prob)
```

```python
accuracy=model.score(x_test,y_test)
print(accuracy)
```

```python
plt.scatter(df['age'],df['bought_insurance'])
```

```python
age_range = np.linspace(df.age.min(),df.age.max(),300)
age_range_df = pd.DataFrame(age_range,columns=['age'])
```

```python
prob_curve=model.predict_proba(age_range_df)[:,1]
plt.plot(age_range_df,prob_curve)
```

```python
plt.axhline(0.5)
plt.xlabel('age')
plt.ylabel('probobility buy insurance')
plt.title('Logistic Regression')
plt.show()
```
