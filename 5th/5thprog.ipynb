```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('insurance_data.csv')
```

```python
x = df[['age']]
y = df['bought_insurance']
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    x, y , test_size=0.2 , random_state=50
)
```

```python
x
model =LogisticRegression()
model.fit(X_train,y_train)
X_train
```

```python
pred = model.predict(X_test)
print("Prediction classes \n",pred)
```

```python
prob = model.predict_proba(X_test)
print("Prediction probabilities \n",prob)
```

```python
accuracy = model.score(X_test,y_test)
print("model accuracy :",accuracy)
```

```python
plt.scatter(df['age'],df['bought_insurance'])
```

```python
age_range = np.linspace(df.age.min(),df.age.max(),300)
age_range_df = pd.DataFrame(age_range,columns=['age'])
```

```python
prob_curve = model.predict_proba(age_range_df)[:,1]
plt.plot(age_range,prob_curve)
```

```python
plt.axhline(0.5)
plt.xlabel("age")
plt.ylabel("probabilityof buying insurance")
plt.title("Logistic Regression Curve")
plt.show()
```
