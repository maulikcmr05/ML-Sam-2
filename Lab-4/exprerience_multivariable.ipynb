```python
from word2number import w2n
import math

df1 = pd.read_csv('salary_predict_dataset.csv')

# fill missing experience
df1.experience = df1.experience.fillna("zero")

# convert words to numbers
df1.experience = df1.experience.apply(w2n.word_to_num)

# fill test_score
median_test_score = math.floor(df1['test_score'].mean())
df1['test_score'] = df1['test_score'].fillna(median_test_score)

# fill interview_score
median_interview_score = df1['interview_score'].median()
df1['interview_score'] = df1['interview_score'].fillna(median_interview_score)

# prediction
p = reg.predict(df1[['experience','test_score','interview_score']])

df1['Salary'] = p

df1.to_csv('predictions.csv', index=False)

df1
```
