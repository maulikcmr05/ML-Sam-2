### MinMaxScaler Formula

The formula for Min-Max Scaling (Normalization) is:

$$x_{scaled} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

Where:
*   **$x_{scaled}$**: The rescaled value (typically between 0 and 1)
*   **$x$**: The original value
*   **$x_{min}$**: The minimum value in the feature column
*   **$x_{max}$**: The maximum value in the feature column


```python
from sklearn.cluster import KMeans
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('income.csv')
df.head()
```

```python
plt.scatter(df.Age,df['Income($)'])
plt.xlabel('Age')
plt.ylabel('income($)')
plt.show()
```

```python
km = KMeans(n_clusters=3)
y_pred = km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df['cluster'] = y_pred
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green',label='cluster 0')
plt.scatter(df2.Age,df2['Income($)'],color='red',label='cluster 1')
plt.scatter(df3.Age,df3['Income($)'],color='black',label='cluster')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
# plt.scatter(km.cluster_centers_[0,0],km.cluster_centers_[0,1],color='purple',marker='*',label='centroid 0')
# plt.scatter(km.cluster_centers_[1,0],km.cluster_centers_[1,1],color='yellow',marker='*',label='centroid 1')
# plt.scatter(km.cluster_centers_[2,0],km.cluster_centers_[2,1],color='orange',marker='*',label='centroid 2')
plt.xlabel("Age")
plt.ylabel('Income($)')
plt.legend()
plt.show()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df.head()
```

```python
plt.scatter(df.Age,df['Income($)'])
plt.show()
```

```python
km1 = KMeans(n_clusters=3)
y_pred2 = km1.fit_predict(df[['Age','Income($)']])
y_pred2
```

```python
df['cluster'] = y_pred2
df.head()
```

```python
km1.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green',label = 'cluster 0')
plt.scatter(df2.Age,df2['Income($)'],color='red',label = 'cluster 1')
plt.scatter(df3.Age,df3['Income($)'],color='black',label = 'cluster 2')
plt.scatter(km1.cluster_centers_[:,0],km1.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
# plt.scatter(km1.cluster_centers_[0,0],km1.cluster_centers_[0,1],color='purple',marker='*',label='centroid 0')
# plt.scatter(km1.cluster_centers_[1,0],km1.cluster_centers_[1,1],color='yellow',marker='*',label='centroid 1')
# plt.scatter(km1.cluster_centers_[2,0],km1.cluster_centers_[2,1],color='orange',marker='*',label='centroid 2')
plt.xlabel("Age")
plt.ylabel('Income($)')
plt.legend()
plt.show()
```
